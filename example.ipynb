```julia
using WaveSpectra
using Plots: plot
# using CairoMakie
using TimeSeries

using WaveRealizations
```

```julia
f̅ = (0.01:0.01:0.5)*Hz
θ̅ = (0°:10°:360°)[begin:end-1]
Hₛ = 1.2m
fₑ = 1/8s
θₘ = 90°
spread = 20°
γ = 3.3
Sₒ = ParametricSpectra.spectrum_jonswap(f̅, Hₛ, fₑ, γ)
D = ParametricSpectra.spread_cartwright(θ̅, f̅, θₘ, spread)
S = Spectrum(Sₒ, D)
```

```julia
# plot_spectrum(Sₒ)
plot(Sₒ)
```

```julia
# plot_spectrum(S)
plot(S)
```

```julia
S_sub = Spectrum(S[frequency=0Hz..0.2Hz, direction=45°..135°])
# plot_spectrum(S_sub, make_cyclic=false)
plot(S_sub)
```

# Wave realizations

```julia
Ã = ComplexAmplitudes(S_sub)
```

```julia
abs.(Ã.data)
# angle.(°, Ã.data)
```

```julia
time = (0:0.5:20)s
x = y = (-1:0.01:1)*60m
dispersion = DispersionRelations.gravitywaves_deepwater()
η = WaveRealizations.WaveSurface(Ã; x, y, time, dispersion)
```

```julia
plot(η[time=0.5s])
```

```julia
gif = surface_gif(η)
```
